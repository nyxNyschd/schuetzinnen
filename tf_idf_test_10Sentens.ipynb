```
import pandas as pd
import math
from collections import Counter
```

```
corpus = [
"He would only survive if he kept the fire going and he could hear thunder in the distance.",
"The delicious aroma from the kitchen was ruined by cigarette smoke.",
"A quiet house is nice until you are ordered to stay in it for months.",
"He was sure the Devil created red sparkly glitter.",
"The busker hoped that the people passing by would throw money, but they threw tomatoes instead, so he exchanged his hat for a juicer.",
"She tilted her head back and let whip cream stream into her mouth while taking a bath.",
"I love eating toasted cheese and tuna sandwiches.",
"I'm confused: when people ask me what's up, and I point, they groan.",
"They called out her name time and again, but were met with nothing but silence.",
"The fog was so dense even a laser decided it wasn't worth the effort.",
"The lake is a long way from here."
];
```

```
corpus_tf = [];
def compute_tf(text):
    text = text.split()
    tf_text = Counter(text)
    tf_text = {i: tf_text[i]/float(len(text)) for i in tf_text}
    return tf_text

for text in corpus:
    corpus_tf.append(compute_tf(text))

print(corpus_tf)
pd.DataFrame(corpus_tf)
```

```
unic_words = set()
for text in corpus:
    text = text.split()
    unic_words = set(unic_words).union(set(text))
print(unic_words)
```

```
def compute_idf(word, corpus):
    return math.log10(len(corpus)/sum([1.0 for i in corpus if word in i]))
```

```
word_idf = {};
for word in unic_words:
    word_idf[word] = compute_idf(word, corpus)
    
#print(word_idf)
pd.DataFrame([word_idf])
```

```
index = {}
for i, text_tf in enumerate(corpus_tf):
    for word in text_tf.keys():
        if word not in index:
            index[word] = {}
        index[word][i] = text_tf[word]*word_idf[word]
print(index)
```

```
query = 'the'
if query in index:
    for i in index[query].keys():
        print(i, corpus[i])
else:
    print("Es gibt kein Satz mit dem Wort " + query)
```
